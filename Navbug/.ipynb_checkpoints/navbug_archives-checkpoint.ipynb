```python
import numpy as np
import pandas as pd
import requests
from bs4 import BeautifulSoup
import time
```

```python
url = "https://www.navbug.com/massachusetts/medford_1_archived_reports.htm"
res=requests.get(url)
res.status_code
```

```python
soup=BeautifulSoup(res.content, 'lxml')
```

```python
news=soup.find('td',{'style':'padding:4px 4px 4px 4px;'}).find_all('a')
```

```python
l2=[]
for i in news[:5]: # in the site there is data for 38 month. so you can enter max as 38 here
    time.sleep(1)
    url=i.attrs['href']
    res=requests.get(url)
    sub_soup=BeautifulSoup(res.content, 'lxml')
    tables=sub_soup.find_all('table',{'class':'listarticle'})

    for idx in enumerate(tables):
        c=idx[0]
#     l2=[]
    for t in range(c):
        bos2={}
        try:
            subtitles=tables[t].find('div',{'class':'listarticle_subtitles'}).text.replace('|',',').strip('\n').rstrip(',')
            story=tables[t].find('div',{'class':'listarticle_body'}).text
    #         print(subtitles)
        except:
            subtitles='No information'

        bos2['month']=i.text
        bos2['news_feed']=story
        bos2['title']=subtitles
        l2.append(bos2)
    
```

```python
l2
```

```python
df=pd.DataFrame(l2)
df.head()
```

```python
city_table=pd.read_table('./data/city list.txt')
city_table['City']=city_table['City'].map(lambda x: x.upper())
city_table.head()
```

```python
city_list=list(city_table['City'])
```

```python
df['title']=df['title'].map(lambda x: x.upper())
```

```python
df['title']=df['title'].map(lambda x: x.replace('MASSACHUSETTS',''))
```

```python
df['city']=''
df['road']=''
```

```python
df.head()
```

```python
for i in df.index:
    words=df['title'][i].split(',')
    for j in words:
        j=j.replace(" ","")
        if j in city_list:
            df['city'][i]=j
        elif j==' ':
            pass
        else:
            df['road'][i]=j
    
```

```python
df
```

```python
df=df[['month','city','road','news_feed']]
df.head()
```
